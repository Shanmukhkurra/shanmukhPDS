```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('StudentsPerformance.csv')

# Countplot for gender
df['gender'].value_counts().plot(kind='bar')
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig('./plots/gender_count.png')
plt.show()
```

```python
# Histogram for math scores
df['math score'].plot(kind='hist', edgecolor='black')
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
plt.savefig('./plots/math_score_distribution.png')
plt.show()
```

```python
# Group by parental level of education and find average reading scores
avg_scores = df.groupby('parental level of education')['reading score'].mean()
# Line plot for average reading scores by parental level of education
avg_scores.plot(kind='line', marker='o')
plt.title('Average Reading Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Reading Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('./plots/avg_reading_scores.png')
plt.show()
```

```python
# Pie chart for race/ethnicity
df['race/ethnicity'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Race/Ethnicity')
plt.savefig('./plots/race_ethnicity_distribution.png')
plt.show()
```

```python
# Scatter plot for math vs reading scores
df.plot(kind='scatter', x='math score', y='reading score')
plt.title('Scatter Plot of Math Scores vs Reading Scores')
plt.savefig('./plots/math_vs_reading_scores.png')
plt.show()
```
